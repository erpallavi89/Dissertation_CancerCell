```python
#!/usr/bin/python
import os
import numpy as np
import scipy.io as sio
```

```python
"""
    To make each image of same dimension, 
    padded all image with zeroes and made each image of dimension (140X140X389)
    and saved it into new directory.
"""

dataDir = "01_Barrett_TMA_MAT/"
newDir="New_Barrett_MatFiles/"
data_x ={}
for file in os.listdir( dataDir ) :
    if file.endswith('.mat'):
        x_pad=[]
        print(dataDir+file)
        data_x=sio.loadmat(dataDir+file)
        shp=data_x['spcImage'].shape
        img=data_x['spcImage']
        leftPad = round(float((140 - shp[0])) / 2)
        rightPad = round(float(140 - shp[0]) - leftPad)
        topPad = round(float((140 - shp[1])) / 2)
        bottomPad = round(float(140 - shp[1]) - topPad)
        pads = ((leftPad,rightPad),(topPad,bottomPad),(0,0))
        for i,x in list(enumerate(img)):
            cons = np.int(np.median(x))
            x_pad= np.pad(img,pads,'constant',constant_values=cons)#pad with zero value
        data_x['spcImage'] = x_pad
        sio.savemat(newDir+file,data_x) #save the new mat file
        print(newDir+file)
        


        
```
